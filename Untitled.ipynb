```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(is_float)]
```

```python
def range_to_num(x):
    nums = x.split('-')
    if len(nums) == 2:
        return (float(nums[0]) + float(nums[1])) / 2
    try:
        return float(x)
    except:
        return None
```

### Generalize the total square feet column value

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(range_to_num)
df4
```

## Feature Engineering

```python
df5 = df4.copy()
df5['price_per_sqft'] = df5['price'] / df5['total_sqft']
df5.head()
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: x.strip())

location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending = False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

### Categorize the categories with few data points as the remnant

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_stats[location_stats<=10] else x)
```

### Remove outliers which have price_per_sqft dramatic

```python
def remove_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7 = remove_outliers(df6)
df7.shape
```

```python
def plot_scatter_chart(df,location):
    bhk2 = df[(df.location==location) & (df.bhk==2)]
    bhk3 = df[(df.location==location) & (df.bhk==3)]
    matplotlib.rcParams['figure.figsize'] = (25,20)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(location)
    plt.legend()
    
plot_scatter_chart(df7,"Rajaji Nagar")
```

```python
def remove_bhk_outliers(df):
    bhk_removals = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean' : np.mean(bhk_df.price_per_sqft),
                'std' : np.std(bhk_df.price_per_sqft),
                'count' : bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stat = bhk_stats.get(bhk-1)
            if stat and stat['count'] > 5:
                bhk_removals=np.append(bhk_removals,bhk_df[bhk_df.price_per_sqft < stat['mean']].index.values)
    return df.drop(bhk_removals, axis='index')
df8 = remove_bhk_outliers(df7)
# df8 = df7.copy()
df8.shape
```

## Machine learning model

```python
df10 = df9.drop(['size','price_per_sqft'],axis='columns')
df10.head(10)
```

```python
dummies = pd.get_dummies(df10.location)
```

```python
df11 = pd.concat([df10,dummies.drop('other',axis='columns')],axis='columns')
```

```python
df11
```

```python
df12 = df11.drop('location',axis='columns')
df12
```

```python
df12.shape
```

```python
X = df12.drop('price',axis = 'columns')
X
```

```python
y = df12.price
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
LinReg.fit(X_train,y_train)
LinReg.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5,test_size = 0.2, random_state=0)
cv
```

```python
cross_val_score(LinearRegression(),X,y,cv=cv)
```
